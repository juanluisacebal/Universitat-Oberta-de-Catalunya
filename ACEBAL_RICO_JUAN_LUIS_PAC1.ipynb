# **2.3.Función clean**

```python
# Funcion clean que acepta un DataFrame, el nombre de una columna, un diccionari
# con las sutituciones y una función de conversión como parámetros,
# y devuelve un DataFrame modificado.
def clean(df: pd.DataFrame, columna: str,
          dicc: dict, funcion_conversion=None) -> pd.DataFrame:
     # Itera sobre cada clave-valor en el diccionario de reemplazos.
    for clave, valor in dicc.items():
        # Reemplaza en la columna especificada del DataFrame todas
        #las ocurrencias de la clave por el valor.
        df[columna] = df[columna].replace(clave, valor, regex=True)
        # Comprueba si he pasado la función de conversión o no
        # Si la hay modifica el valor con la función de conversión
        # Si no, devuelvo el df
        if funcion_conversion is not None:
    # Aplico la función de conversión a cada elemento de la columna del DataFrame,
    # lo que modifico los valores según la función de conversión.
          df[columna] = df[columna].apply(funcion_conversion)
    # Devuelve el DataFrame modificado después de realizar todos los cambios.
    return df

# Definición de la función convertir_a_megabytes que acepta un valor en forma
# de cadena y devuelve una cadena,
def convertir_a_megabytes(valor: str) -> str:
    # Busca un patrón que coincida con un número opcionalmente seguido de "M" o "k" al final del valor.
    # https://regexper.com/#%5E%28%5Cd%2B%28%5C.%5Cd%2B%29%3F%29%5Cs*%28%5BMk%5D%29%24
    match = re.match(r'^(\d+(\.\d+)?)\s*([Mk])$', valor)
    # Si se encuentra una coincidencia con el patrón definido.
    if match:
        # Extrae el número encontrado en el valor.
        numero = float(match.group(1))
        # Extrae la unidad de medida ("M" para megabytes, "k" para kilobytes)
        unidad = match.group(3)
        # Si la unidad es "M" (megabytes), lo dejo igual
        if unidad == 'M':
            return numero
        # Si la unidad es "k" (kilobytes), realizo la conversión multiplicando
        # el número por 1000 y formatea el resultado con 4 decimales.
        elif unidad == 'k':
            return "{:.4f}".format(numero / 1000)
    # Si no se encuentra una coincidencia con el patrón definido, devuelve el
    # valor original sin cambios.
    return valor
```

# **3.1. Definir funcion change_format**

```python
def change_format(cad: str) -> str:
    # Busca una coincidencia con el patrón en la cadena
    match = re.match(r'^(\d+)--(\w+)--(\d+)$', cad)
    if match:
        # Extrae el día, mes y año de la cadena
        day, month, year = match.groups()
        # Mapea los nombres de los meses a números
        months = {'Jan': '1', 'Feb': '2', 'Mar': '3', 'Apr': '4', 'May': '5', 'Jun': '6',
                  'Jul': '7', 'Aug': '8', 'Sep': '9', 'Oct': '10', 'Nov': '11', 'Dec': '12'}
        # Devuelve la cadena en el formato dd-mm-yyyy
        return f"{day}-{months[month]}-{year}"
    else:
        return cad  # Si no se encuentra un formato válido, devuelvo la cadena original
```

# **3.2. Definir funcion convert_dat**

```python
def convert_date(dataframe: pd.DataFrame) -> pd.DataFrame:
    # Aplica la función change_format a la columna Last_Updated
    dataframe['Last_Updated'] = dataframe['Last_Updated'].apply(change_format)
    dataframe['Last_Updated'] = pd.to_datetime(dataframe['Last_Updated'],
                                               format='%d-%m-%y')#, errors='coerce')
    return dataframe
```
