````python
datag = data.groupby('Location')
for loc, i in datag:
    caracteristicas_compra = loc
    cantidad_media_compra = i['Purchase Amount (USD)'].mean()
    categoria_mas_popular = i['Category'].value_counts().idxmax()
    genero_mas_frecuente = i['Gender'].value_counts().idxmax()
    edad_media_comprador = i['Age'].mean()
    print(f"Características de compra en {caracteristicas_compra}.")
    print(f"Cantidad media de compra: {cantidad_media_compra}.")
    print(f"Categoría más popular: {categoria_mas_popular}.")
    print(f"Género del comprador más frecuente: {genero_mas_frecuente}.")
    print(f"Edad media del comprador: {edad_media_comprador}.\n")
````

Propón una gráfica de sector que sea representativa a tu elección. Se valorará la complejidad.

````python

# Calcular las 10 ciudades principales en facturación
top_20_ciudades = data.groupby('Location')['Purchase Amount (USD)'].sum().nlargest(25)

# Sumar la facturación del resto de las ciudades
resto_facturacion = data.groupby('Location')['Purchase Amount (USD)'].sum().sum() - top_20_ciudades.sum()

# Crear una lista de colores para las 10 ciudades principales y el resto en negro
colores = list(plt.cm.tab10.colors[:len(top_20_ciudades)]) + ['black']

# Crear etiquetas para las 10 ciudades principales y el resto
etiquetas = list(top_20_ciudades.index) + ['Resto']

# Calcular los tamaños de las porciones del gráfico
tamaños = list(top_20_ciudades.values) + [resto_facturacion]

# Crear el gráfico de sector
plt.figure(figsize=(11, 8))
plt.pie(tamaños, labels=etiquetas, autopct='%1.1f%%', colors=colores, startangle=90)

# Añadir título
plt.title('Distribución de facturación por ciudad (Top 10 y Resto)')

# Mostrar el gráfico
plt.show()
````

El [protocolo Open Graph ](https://ogp.me/) es un estándar desarrollado por Facebook que permite que cualquier página web tenga la misma funcionalidad que otro objeto en Facebook.

Para hacer esto posible es necesario añadir metadatos en la página, es decir, añadir una serie de tags `<meta>` en la parte `<head>` de la página web.

```
<html prefix="og: https://ogp.me/ns#">
<head>
<title>The Rock (1996)</title>
<meta property="og:title" content="The Rock" />
<meta property="og:type" content="video.movie" />
<meta property="og:url" content="https://www.imdb.com/title/tt0117500/" />
<meta property="og:image" content="https://ia.media-imdb.com/images/rock.jpg" />
...
</head>
...
</html>

```

En esta parte vamos a extraer metadatos open graph de una página web, en concreto, de un vídeo de YouTube.

````python
import requests
from bs4 import BeautifulSoup
````

El código a continuación obtiene el código fuente de la página web, tal y como lo hace un navegador, y lo carga en un objeto BeautifulSoup desde el que podemos extraer los metadatos.

````python
url = 'https://es.wikipedia.org/wiki/The_Game_Awards' # dada la siguiente url de la wikipedia

headers = { }
r = requests.get (url=url, headers=headers)

soup = BeautifulSoup (r.text, 'html.parser')
````

Mediante los métodos `find()`o  `find_all()` es posible extraer los datos

*   Elemento de lista
*   Elemento de lista

de la página.

````python
# Recuperar todos los tag meta
soup.find_all ('meta')
````

````python
# Obtener todos los tag meta con la propiedad og:title
soup.find ("meta", property="og:title")
````

````python
# TODO [1] : obtener la descripción open graph

og_descripcion = soup.find("meta", property="og:description")
if og_descripcion:
    og_descripcion = og_descripcion.get("content")
else:
    og_descripcion = " no encontrada"

print("Descripción Open Graph: ", og_descripcion)
````

````python
# TODO [2] : obtener la imagen open graph
og_imagen = soup.find("meta", property="og:image")
if og_imagen:
    og_imagen = og_imagen.get("content")
else:
    og_imagen = "Imagen Open Graph no encontrada"

print("Imagen Open Graph:", og_imagen)
````

Obtener todos los los tags openg graph y almacernarlos en un diccionario
```
  { 'og:title': ... }
```


````python
# TODO [3]
og_tags = {}
og_meta_tags = soup.find_all("meta", property=lambda x: x and x.startswith("og:"))
for tag in og_meta_tags:
    property_name = tag.get("property")
    content = tag.get("content")
    if property_name and content:
        og_tags[property_name] = content

print("Tags Open Graph:")
for key, value in og_tags.items():
    print(f"{key}: {value}")
````
