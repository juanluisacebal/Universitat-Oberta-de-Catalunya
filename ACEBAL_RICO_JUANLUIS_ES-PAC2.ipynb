# Ejercicio 1

Queremos crear una función que compruebe si un año es un año bisiesto o no.

La función recibirá dos parámetros:
- el año que queremos comprobar,
- un parámetro **opcional** que nos indique si queremos que la función muestre por pantalla la razón por la cual el año no se considera un año bisiesto (en el caso que el año sea bisiesto, la función no deberá mostrar ningún mensaje de todas formas).

La función devolverá un dato de tipo booleano (True / False)*.

Según el calendario Gregoriano, un año se considera año bisiesto si cumple las siguientes condiciones:
- Si 4 es un divisor del año, es un año bisiesto, excepto
 - Si 100 también es un divisor del año, en cuyo caso el año **NO** se considera bisiesto, excepto
   - Si 400 también es un divisor del año, en cuyo caso el año **SÍ** se considera bisiesto.
   
Por ejemplo, a continuación podéis ver algunos ejemplos del resultado esperado en función del año.

2000 -> True

1800 -> False

2100 -> False

**Nota**. **No podéis utilizar la función `isleap` que hemos visto en la Unidad 1**. Aplicad la función implementada a los años indicados en la celda inferior y comprobad que el resultado es el mismo que el resultado obtenido si utilizamos la función `isleap` de Python.

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

```python
# Respuesta
import calendar


def es_bisiesto(año: int, mostrar_mensaje: bool = False) -> bool:
    """
    Determina si un año es bisiesto.

    Args:
        año (int): El año a comprobar.
        mostrar_mensaje (bool, opcional): Si es True, muestra el motivo.

    Returns:
        bool: True si el año es bisiesto, False en caso contrario.
    """
    if año % 4 == 0:
        if año % 100 == 0:
            if año % 400 == 0:
                if mostrar_mensaje:
                    print("El año {} -> BISIESTO. "
                          "Divisible entre 400".format(año))
                return True
            else:
                if mostrar_mensaje:
                    print("El año {} -> NO Bisiesto. divisible entre 100, "
                          "pero no entre 400.".format(año))
                return False
        else:
            if mostrar_mensaje:
                print("El año {} -> BISIESTO. Divisible entre 4, pero"
                      " no entre 100".format(año))
            return True
    else:
        if mostrar_mensaje:
            print("El año {} -> NO Bisiesto."
                  "No divisible entre 4.".format(año))
        return False
```

```python
years = [1800, 1996, 2000, 2101, 2400]
```

```python
for year in years:
    custom_leap = es_bisiesto(year)
    python_leap = calendar.isleap(year)
    print(f"Año: {year} -> Mi funcion: {custom_leap}, isleap: {python_leap} ")
    assert custom_leap == python_leap, "La implementación difiere del estándar"
```

```python
for year in years:
    print(f"Año: {year} -> Mi funcion: {es_bisiesto(year, True)} \n ")
```

```python
# Probando la función con algunos años
years = [1800, 1996, 2000, 2101, 2400]
results = {year: es_bisiesto(year, True) for year in years}
print(results)
```

```python
import pandas as pd
import re


def create_dictionary(filename: str, columna: str) -> dict:
    """
    Crea un diccionario agrupando jugadores por
    código de país extraído de la URL de la bandera.

    Args:
        filename (str): Ruta al CSV que contiene los datos.
        columna (str): Nombre de la columna de la que se
        trataran los datos.

    Returns:
        dict: Dicc con claves de codigos de países y valores
        como listas de tuplas (nombre corto, valor columna).
    """
    df = pd.read_csv(filename)
    jugadores_por_nacion = {}

    for index, fila in df.iterrows():
        url_bandera = fila['nation_flag_url']
        # Extrae el codigo de pais de la URL de la bandera
        # La expresion regular busca aa.png o bien aa-bbb.png
        # Donde aa es es por ejemplo y bbb es para gb-eng por ejemplo
        match = re.search(r'flags/([a-z]{2}(?:-[a-z]{3})?)\.png', url_bandera)
        if match:
            # Código de región,'es','gb-eng', etc
            codigo_pais = match.group(1)
            # Si la clave no existe
            if codigo_pais not in jugadores_por_nacion:
                jugadores_por_nacion[codigo_pais] = []
            # Tuplo con el nombre corto, y valor de la columna pasada
            # Si es nulo se deja la cadena vacia
            informacion_jugador = (fila['short_name'],
                                   str(int(fila[columna]))
                                   if pd.notna(fila[columna]) else "")
            # agrego el jugador
            jugadores_por_nacion[codigo_pais].append(informacion_jugador)

    return jugadores_por_nacion
```

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

```python
import zipfile
import os


def descompresion_zip(zip_path: str, extract_folder: str):
    """
    Descomprime un archivo ZIP en la carpeta especificada.

    Args:
    zip_path (str): La ruta completa al archivo ZIP.
    extract_folder (str): Nombre carpeta donde se descomprimirán los archivos.
    """
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_folder)
    print(f"Archivos descomprimidos en la carpeta: {extract_folder}")
```

```python
def tamano_archivo(file_path: str) -> float:
    """
    Calcula el tamaño de un archivo en megabytes.

    Args:
    file_path (str): La ruta al archivo.

    Returns:
    float: El tamaño del archivo en megabytes.
    """
    size = os.path.getsize(file_path) / (1024 * 1024)
    return size


def mostrar_tamano_archivos(*file_paths: str):
    """
    Muestra el tamaño de varios archivos.

    Args:
    *file_paths (str): Rutas a los archivos cuyos tamaños se desean calcular.
    """
    for file_path in file_paths:
        size = tamano_archivo(file_path)
        print(f"El tamaño de '{os.path.basename(file_path)}' es {size:.2f} MB")
```

**Notas importantes:**

2) Como parte del ejercicio, deberá separarse el texto en frases. Éste es un tema que puede llegar a ser muy complejo y por tanto en este ejercicio trabajaremos con una simplificación. En concreto, consideraremos que las frases terminarán **siempre** con un punto "." o con un símbolo de interrogación "?". Adicionalmente, el texto ha sido simplificado para no contener algunos símbolos conflictivos como los puntos suspensivos. Nótese que muchas de las frases están repartidas en más de una línea de texto. Las frases que se devuelven deben incluir, en todo caso, el símbolo de final de frase (punto "." o "?").

3) Utilizad los principios de **programación funcional** que hemos visto en teoría para resolver este ejercicio y **expresiones regulares**.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(3 puntos)**

```python
def procesar_libro(ruta_archivo: str, patron: str) -> list:
    """
    Procesa un libro y encuentra las frases que coincidan con un patrón dado.

    Args:
        ruta_archivo (str): La ruta al archivo de texto.
        patron (str): La expresión regular a buscar dentro del archivo.

    Returns:
        list: Una lista de frases que contienen la coincidencia.
    """
    with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
        texto = archivo.read()
        # Dividimos el texto en frases
        frases = re.split(r'\.|\?', texto)
        # Aseguramos que cada frase termina con su puntuación final
        frases_con_puntuacion = [
            frase + (texto[texto.index(frase) + len(frase)]
                     if texto.index(frase) + len(frase) < len(texto) else '')
            for frase in frases if frase.strip()
        ]
        # Patrón para que no distinga mayúsculas o minusculas
        expresion_regular = re.compile(patron, re.IGNORECASE)
        # Filtramos las frases que contienen el patrón
        frases_coincidentes = [
                            frase.strip('\n') for frase
                            in frases_con_puntuacion
                            if expresion_regular.search(frase)
                            ]
    return frases_coincidentes
```

```python
# Patrón para "War of the Worlds" incluye 'war' o 'worlds'
pattern_worlds = r'\b(war|worlds)\b'
pattern_nothing = r'\b(war|nothing)\b'


# Patrón para "Moby Dick" incluye palabras que empiezan por 'whal'
pattern_whale = r'\bwhal\w*\b'

# Llamadas
war_sentences_nothing = procesar_libro('./Data/warworlds.txt', pattern_nothing)
war_sentences = procesar_libro('./Data/warworlds.txt', pattern_worlds)
whale_sentences = procesar_libro('./Data/mobyDick.txt', pattern_whale)

# Visualización resultados
print("Frases que contienen 'war' o 'worlds':", war_sentences[:5])
print("Frases que contienen palabras"
      " que empiezan por 'whal':", whale_sentences[:5])
```
